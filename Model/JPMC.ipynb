```python
boards = ['CBSE', 'KA', 'MH', 'TS']
subjects = ['EVS', 'English', 'Math']
grades = ['G1', 'G2', 'G3', 'G4', 'G5']
```

```python
code_list = lesson_df["Code"][0].split('-')
```

```python
code_list
```

```python
#Board wise count
board_df = pd.DataFrame({"Board" : boards})
subject_df = pd.DataFrame({"Subject" : subjects})
grade_df = pd.DataFrame({"Grade": grades })
grade_df
```

```python
dates = list(lesson_df.columns)
dates.remove('Type')
dates.remove('Code')
```

```python
dates
```

```python
for d in dates:
    board_df[d] = 0
    grade_df[d] = 0
    subject_df[d] = 0
```

```python
subject_df
```

```python
#Board wise count
def get_board_wise():
    for i in range(lesson_df.shape[0]):
        for j in dates:
            code = lesson_df["Code"][i].split('-')
            cnt = lesson_df[j][i]
            code = code[0]
            index = board_df.index[board_df['Board']==code].tolist()
            board_df.at[index[0], j] += cnt
```

```python
#Grade wise count
def get_grade_wise():
    for i in range(lesson_df.shape[0]):
        for j in dates:
            code = lesson_df["Code"][i].split('-')
            cnt = lesson_df[j][i]
            code = code[2]
            index = grade_df.index[grade_df['Grade']==code].tolist()
            grade_df.at[index[0], j] += cnt
```

```python
#Subjects
def get_subject_wise():
    for i in range(lesson_df.shape[0]):
        for j in dates:
            code = lesson_df["Code"][i].split('-')
            cnt = lesson_df[j][i]
            code = code[1]
            index = subject_df.index[subject_df['Subject']==code].tolist()
            subject_df.at[index[0], j] += cnt
```

```python
get_board_wise()
get_subject_wise()
get_grade_wise()
```

```python
board_df["Total"] = 0
grade_df["Total"] = 0
subject_df["Total"] = 0
```

```python
def get_total():
    for i in range(board_df.shape[0]):
        for j in dates:
            cnt = board_df[j][i]
            board_df['Total'][i] += cnt
    for i in range(grade_df.shape[0]):
        for j in dates:
            cnt = grade_df[j][i]
            grade_df['Total'][i] += cnt
    for i in range(subject_df.shape[0]):
        for j in dates:
            cnt = subject_df[j][i]
            subject_df['Total'][i] += cnt
```
